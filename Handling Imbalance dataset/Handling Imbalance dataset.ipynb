Imbalanced Dataset:
-  A dataset with unequal class distribution

```python
#importing dependencies
import numpy as np
import pandas as pd
```

```python
# collecting dataset
dataset = pd.read_csv("/Users/tushark/Python_datasets/credit_data.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.tail()
```

```python
# distribution of the two classes
dataset['Class'].value_counts()
```

 This is highly imbalanced dataset : 
0==> legit Transactions
1==> Fradulent transaction

```python
# separating the legit and fradulent transactions
legit = dataset[dataset.Class==0]
fradulent = dataset[dataset.Class==1]
```

```python
print(legit.shape)
print(fradulent.shape)
```

Under sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactionds

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fradulent], axis = 0)
```
