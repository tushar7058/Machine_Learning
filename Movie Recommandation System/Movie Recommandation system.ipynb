```python
# import the libraries
import numpy as np
import  pandas as pd
import  difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Data Collection and preprocessing
movies_Data = pd.read_csv("/Users/tushark/Downloads/CSV/movies.csv")
movies_Data.head()
```

```python
movies_Data.shape
```

```python
movies_Data.isnull().sum()
```

```python
# selecting the relevant features for recommendation
select_feature = ['genres','homepage','keywords','overview']
print(select_feature)
```

```python
# replacing the null values with null string
for feature in select_feature:
    movies_Data[feature]= movies_Data[feature].fillna('')
```

```python
# combining all the 5 selected features
combine_feature = movies_Data['genres']+''+movies_Data['homepage']+''+movies_Data['keywords']+''+movies_Data['overview']
print(combine_feature)
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combine_feature)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
# getting the similarity scores using cosine similarity
similarity  = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

Getting the Movie Name from User

```python
# getting the movie name from user 
movies_name = input('Enter your Favourite movie name:')
```

```python
# creating a list with all the movie names given in the dataset
list_of_all_title = movies_Data['original_title'].tolist()
```

```python
print(list_of_all_title)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movies_name,list_of_all_title)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title
index_of_the_movie = movies_Data[movies_Data.original_title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting list of all similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies Suggested for You :\n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_Data[movies_Data.index==index]['original_title'].values[0]
    if i<30:
        print(i,'_',title_from_index)
        i+=1
```

Movie Recommendation System

```python
movies_name = input('Enter Your Favourite Movie Name :\n')

list_of_all_title = movies_Data['original_title'].tolist()

find_close_match = difflib.get_close_matches(movies_name,list_of_all_title)

close_match  = find_close_match[0]

index_of_the_movie = movies_Data[movies_Data.original_title==close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('Movies Suggested For you :\n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_Data[movies_Data.index==index]['original_title'].values[0]
    if i<30:
        print(i,'_',title_from_index)
        i+=1
    
        
```
