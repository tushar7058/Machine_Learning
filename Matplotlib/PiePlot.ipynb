Pieplot in matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(10,30,(20))
print(x) 
```

```python
num = [24, 14, 10, 21, 11, 15, 12, 29, 10, 20, 24, 11, 10, 26, 13, 21, 16 ,15 ,27, 26]
plt.pie(num)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%")
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",radius=1.5)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",labeldistance=.8,shadow=True)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",counterclock=False)
plt.show()
```

```python
x = [10,20,30,40]
y =["c","c++","java","python"]
ex= [0.4,0.0,0.0,0.0]
c = ["r","g","b","y"] 

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",wedgeprops={"linedrops":4})
plt.show()
```
