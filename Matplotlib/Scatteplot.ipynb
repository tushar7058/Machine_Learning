*** Scatter_plot***


```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y =['a','b','c','d','e','f','g','h']
plt.scatter(x,y , edgecolors='yellow')
plt.title('scatterplot',fontsize=15)
plt.xlabel('Alphabets',color='r')
plt.ylabel('Numbers',color='r')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y =['a','b','c','d','e','f','g','h']
colors = ['red','green','yellow','blue','black','pink','black','blue']
sizes = [100,200,300,400,500,600,700,850]
plt.scatter(x,y,s=sizes,alpha=0.8,marker='*',edgecolor='g', c=colors,linewidths=2)
plt.title('scatterplot',fontsize=15)
plt.xlabel('Alphabets',color='r')
plt.ylabel('Numbers',color='r')
plt.show() 
```

```python
x = [1,2,3,4,5,6,7,8]
y =['a','b','c','d','e','f','g','h']
colors = ['red','green','yellow','blue','black','pink','black','blue']
sizes = [100,200,300,400,500,600,700,850]
plt.scatter(x,y,s=sizes,c=colors,cmap='viridis')
plt.colorbar()
plt.title('scatterplot',fontsize=15)
plt.xlabel('Alphabets',color='r')
plt.ylabel('Numbers',color='r')
plt.show() 
```
