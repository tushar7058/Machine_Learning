BarPlot in matplot library

```python
import matplotlib.pyplot as plt
```

```python
x = ['python','java','c++','c']
y = [80,75,60,40]
plt.xlabel('languages')
plt.ylabel('percentage of popularity')
plt.title("language_Data",fontsize=20,color='y',)
plt.bar(x,y,width=0.2,color='y',label = "know")
plt.bar(x,y,label= "popularity")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["sun", "mon", "tue", "wed", "thu", "fri", "sat"]
y = ["1","2","3","4","5","6","7"]

plt.xlabel("Days")
plt.ylabel("Values",fontsize=20)
plt.title("Days vs Values",fontsize=20)
colorc = ["red","green","blue","yellow","orange","purple","black"]
plt.bar(x, y,width=0.5, color=colorc,align='center',edgecolor='black',linewidth=2)
plt.grid()
plt.show()
```
