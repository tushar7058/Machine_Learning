BarPlot in matplot library

```python
import matplotlib.pyplot as plt
```

```python
x = ['python','java','c++','c']
y = [80,75,60,40]
plt.xlabel('languages')
plt.ylabel('percentage of popularity')
plt.title("language_Data",fontsize=20,color='y',)
plt.bar(x,y,width=0.2,color='y',label = "pupularity")
plt.bar(x,y,label= "popularity")
plt.show()
```
